* **Comparison:** `== != > < >= <=`

* **Membership:** `in, not in`
* **Identity:** `is, is not`

```python
print(4+5)
print(4-5)
print(4*5) 
print(4/5)  # Division
print(1//2) # Floor Division
print(2**2) # Exponentiation
print(5%4)  # Modulus the remainder will be the answer not the actual answer
print(4 == 5) # Equal
print(1 != 5) # Not Equal
print(4 > 5)  # Greater than
print(4 < 5)  # Less tha
print(4 >= 5) # Greater than or equal to
print(4 <= 5) # Less than or equal to
```

#### ðŸ”¹ 5. Python Sets

**Notes:**

* Set is unordered, unindexed, no duplicates.

```python
my_set = {"apple", "banana", "cherry", "banana"}

print(my_set)
```

```python
print(my_set[0])
```


##### Add & Remove

```python
my_set = {"apple", "banana", "cherry", "banana"}

my_set.add("mango")
my_set.remove("banana")

print(my_set)
```

##### Set Operations

```python
a = {1,2,3}
b = {3,4,5}
print(a.union(b))      # {1,2,3,4,5}
print(a.intersection(b))  # {3}
print(a.difference(b))
```

```python
myfamily = {"child1": {"name":"Ade", "year":2004}, "child2": {"name":"punk", "year":2007}, "child3": {"name":"fathia", "year":2011}}
myfamily
```

```python
# access the nest dict
# print(myfamily['child2']['name'])

# loop through a nested dict by using the item() method
for x, y in myfamily.items():
    print(x, y)
    for i in y:
        print(i, y[i] )
```
