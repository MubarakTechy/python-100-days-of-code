üß© Example 1: Simple Match

Exercise:
Write a Python program that uses match-case to print a message depending on a fruit name entered by the user.

Rules:

If the user enters "apple", print "Apples are red or green!"

If the user enters "banana", print "Bananas are yellow!"

If the user enters "orange", print "Oranges are orange!"

Otherwise, print "I don't know that fruit."

````python
favorite_fruits = input("Enter your favorite fruits?").lower()

match favorite_fruits:
    case "apple":
        print("Apples are red or green!")
    case "orange":
        print("Oranges are orange!")    
    case "banana":
        print("banana is yellow")
    case _:
        print("I don't know that fruit.")
````

## üß© Example 2: Matching Multiple Values (Revised)

Exercise:
Write a Python program that takes a school subject as input and prints the type of subject it is.

Rules:

If the subject is "Math" or "Physics", print "This is a Science subject."

If the subject is "History" or "Geography", print "This is a Social Studies subject."

If the subject is "English" or "French", print "This is a Language subject."

Otherwise, print "Unknown subject category."

````python
subject = input("Enter  your subject in school")

match subject:
    case "math" | "physics":
        print("This is a Science subject.")
    case "history" | "geography":
        print("This is a Social Studies subject.")
    case "English" | "french":
        print("This is a Language subject.")
    case _:
        print("Unknown subject category.")
````
